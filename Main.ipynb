```python
# Load dataset
df = pd.read_csv('dataset/ETFs.csv', index_col=False)
```

```python
# Counts for each category
df['count'] = 1

for column in df.columns:
    if column == "count":
        continue
    print(df.groupby(column).count()['count'])
    print()
```

```python
# Distributions
distributions = [
    'category',
    'fund_family',
]

for dis in distributions:
    print()
    print(dis)
    print(df[dis].describe())
    sub_df = df.groupby(dis).count()['count'].plot(kind='hist')
    plt.show()
```

```python
#TODO: Change check to scatter, and then rest of plot types
check = [
    'size',
    'net_assets',
    'ytd_return',
    'investment',
]

# Scatter
for x in range(len(check)):
    for y in range(x + 1, len(check)):
        sns.scatterplot(x=check[y], y=check[x], data=df)
        plt.show()
```

```python
# Line
for x in range(len(check)):
    for y in range(x + 1, len(check)):
        sns.lineplot(x=check[y], y=check[x], data=df)
        plt.show()
```

```python
# Histogram
for x in range(len(check)):
    sns.distplot(df[check[x]], bins=10, kde=True)
    plt.show()
    
```

```python
# Bar
for x in range(len(check)):
    sns.countplot(df[check[x]], bins=10, kde=True)
    plt.show()
```

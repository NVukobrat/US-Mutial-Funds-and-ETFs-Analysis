```python
import os

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Change current working dir
print("Original working dir: ", os.getcwd())
os.chdir(os.path.join(os.getcwd(), "us-mutual-funds-and-etfs-analysis"))
print("New working dir: ", os.getcwd())
```

```python
# Load dataset
df = pd.read_csv('dataset/ETFs.csv')

# Basic info
df['count'] = 1

for column in df.columns:
    print(df.groupby(column).count()['count'])
    print()

analysis_columns = [
    'category',
    'fund_family',
    'investment',
    'size',
    'currency',
]
for i, column in analysis_columns:     
    # Aggregate
    sub_df = df.groupby(column).count()['count']
    
    # Plot
    sub_df.plot.bar()
    plt.show()
    print()
# print(df.describe())


```

```python
# Analysis
sns.pairplot(df, kind="scatter", palette="Set2")
plt.show()
```

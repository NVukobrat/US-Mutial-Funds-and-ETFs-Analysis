```python
# Load dataset
df = pd.read_csv('dataset/ETFs.csv', index_col=False)
```

```python
# Counts for each category
df['count'] = 1

for column in df.columns:
    if column == "count":
        continue
    print(df.groupby(column).count()['count'])
    print()
```

```python
# Distributions
distributions = [
    'category',
    'fund_family',
    'net_assets',
    'ytd_return',
    'fund_yield',
    'legal_type',
    'investment',
    'size',
    'currency',
]

for dis in distributions:
    print()
    print(dis)
    print(df[dis].describe())
    df.groupby(dis).count()['count'].plot(kind='hist')
    plt.show()
```

# Basic analysis

```python
# Populate empty
df.replace('', 'NaN', inplace=True)
df.replace('nan', 'NaN', inplace=True)
df.replace('NaN', np.nan, inplace=True)
df = df.fillna(df.mean)
```

```python
# Count(Bar) plots
count = [
    'legal_type',
    'investment',
    'size',
    'currency',
]
for c in count:
    chart = sns.countplot(df[c])
    chart.set_xticklabels(
        chart.get_xticklabels(),
        rotation=30,
        horizontalalignment='right',
    )
    plt.show()
```

```python
# Encode labels
encode = count.copy()
encode += [
    'category',
    'fund_family',
]

le = LabelEncoder()
for e in encode:
    df[e] = le.fit_transform(df[e])
df.replace('NaN', np.nan, inplace=True)
df = df.fillna(df.mean)
```

```python
# Distribution/Histogram plots
distribution = encode.copy()
for d in distribution:
    sns.distplot(df[d])
    plt.show()
```
